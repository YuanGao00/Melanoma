```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from scipy import stats
```

# Patient Subcutaneous Fat Area & Thickness

image source of the study: https://wiki.cancerimagingarchive.net/pages/viewpage.action?pageId=37225348

We selected male patient PET-CT scan images and collected their subcutaneous fat (sub-q fat) area and subcutaneous fat thickness using slices at the L3-L4 vertebral level. All the measurement are done by using ImageJ.

For each patient's slice, the subcutaneous fat area was measured 3 times and then we took the mean value for later analysis. As for subcutaneous fat thickness, we measured the thickness 3 times for both left and right body sides and then we took the mean value for the analysis.

column meaning in df_subqFat:  
    - 'No.' is the generated patient id  
    - 'area_1' to 'area_3' are the 3 meantured sub-q fat area  
    - 'mean_area' is the mean sub-q area fat for each patient  
    - 'thickness1_R' to 'thickness3_R' are the 3 meantured sub-q fat thickness for right side of the body  
    - 'thickness1_L' to 'thickness3_L' are the 3 meantured sub-q fat thickness for left side of the body  
    - 'mean_thick' is the mean sub-q fat thickness value for each patient  
     

```python
path = '/Users/yuangao/Desktop/research/TZD drug'
file_name = 'filtered_fat'
df_subqFat = pd.read_excel(f'{path}/{file_name}.xlsx', engine='openpyxl')
df_subqFat
```

```python
fat_filtered = df_subqFat[['No.', 'mean_area', 'mean_thick']]
fat_filtered.head()
```

# Patient Pre-treatment and Post-treatment Tumor Size

We also collected the pre-treatment and post-treatment tumor sizes for each patient, denoted as "pre_area" and "after_area" in the df_tumor table, respectively. We selected the axial CT slice with the largest axial area of tumor to represent the tumor size.

For patients without a tumor (e.g., pulmonary edema) or who only had one series of CT scans but could not determine whether they were pre-treatment or post-treatment, we labeled them "NaN." After filtering 42 patients, we have 21 patients left, which were the subjects of our study.

```python
file_name = 'tumor_filter'
df_tumor = pd.read_excel(f'{path}/{file_name}.xlsx', engine='openpyxl')
df_tumor
```

```python
tumor_filtered = df_tumor[(df_tumor['pre_area'].notna()) & (df_tumor['after_area'].notna())].reset_index(drop = True)
tumor_filtered
```

# Combining the two tables and calculating treatment responsiveness

After merging the above 2 tables by maching the patient id, we calculated the treatment responsiveness by using the pre-treatment tumor size to minus the post-treatment tumor size and set the column name as 'response'.

In this part, we define the obesity by using either the sub-q fat area or the sub-q fat thickness. We defined a patient as obese if the thickness was greater than 20 mm and vice versa. In addition, we defined a patient as obese if the area was greater than 25000 mm^2 and vice versa.

```python
df = pd.merge(fat_filtered, tumor_filtered, on = 'No.', how = 'right')
df.rename(columns = {'mean_area': 'fat_area', 
                     'mean_thick': 'fat_thickness'}, 
          inplace=True)
df
```

```python
df['response'] = df['pre_area'] - df['after_area'] 
df['obese_thickness'] = df['fat_thickness'].apply(lambda x: 'obese' if x > 20 else 'non_obese')
df['obese_area'] = df['fat_area'].apply(lambda x: 'obese' if x > 25000 else 'non_obese')
df = df[df['No.'] != 35].reset_index(drop = True)
df

# area_unit: mm^2
# thickness: mm 
```

```python
print(df['obese_area'].value_counts())
print('\n')
print(df['obese_thickness'].value_counts())
```

### t-test

```python
dfx = df.copy()
# Sample data for obese and non-obese groups
obese_group = np.array(dfx[dfx['obese_thickness'] == 'obese']['response']) 
non_obese_group = np.array(dfx[dfx['obese_thickness'] == 'non_obese']['response']) 
# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(obese_group, non_obese_group)

# Print results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Check for significance
alpha = 0.1
if p_value < alpha:
    print("The difference between the groups is statistically significant.")
else:
    print("There is no significant difference between the groups.")
```
